```python
#TASK 1:There are three different tables as given above. Please make three dataframe in python and save them as three .csv files.From Task 2 to Task 10, use the saved .csv files only.

import pandas as pd

#Creating Employee DataFrame
employee_data = {
    "ID": ["A001", "A002", "A003", "A004", "A005"],
    "Name": ["John Alter", "Alice Luxumberg", "Tom Sabestine", "Nina Adgra", "Amy Johny"],
    "Gender": ["M", "F", "M", "F", "F"],
    "City": ["Paris", "London", "Berlin", "Newyork", "Madrid"],
    "Age": [25, 27, 29, 31, 30]}
df_employee = pd.DataFrame(employee_data) #Employee Data is saved in df_employee
df_employee.to_csv("employee_data.csv", index=False) #Converting Data frame into CSV File

# Creating Seniority Level DataFrame
seniority_data = {
    "ID": ["A001", "A002", "A003", "A004", "A005"],
    "Designation Level": [2, 2, 3, 2, 3]
}
df_seniority = pd.DataFrame(seniority_data) #Seniority Data is saved in df_seniority
df_seniority.to_csv("seniority_data.csv", index=False) #Converting Data frame into CSV File

# Creating Project DataFrame
project_data = {
    "ID": ["A001", "A002", "A003", "A004", "A005", "A002", "A005", "A003", "A001", "A003", "A001", "A004", "A004", "A005"],
    "Project": ["Project 1", "Project 2", "Project 3", "Project 4", "Project 5", "Project 6", "Project 7", "Project 8", "Project 9",
                "Project 10", "Project 11", "Project 12", "Project 13", "Project 14"],
    "Cost": [1002000, 2000000, 4500000, 5500000, None, 680000, 400000, 350000, None, 300000, 2000000, 1000000, 3000000, 200000],
    "Status": ["Finished", "Ongoing", "Finished", "Ongoing", "Finished", "Failed", "Finished", "Failed", "Ongoing", "Finished",
               "Failed", "Ongoing", "Finished", "Finished"]
}
df_project = pd.DataFrame(project_data) #Project Data is saved in df_project
df_project.to_csv("project_data.csv", index=False) #Converting Data frame into CSV File

print("Task 1 Completed: Files saved successfully!")
```

```python
#Task 2: The cost column in the dataframe “Project” has some missing values. Your task is to compute these missing values. Replace the missing values by running average. You should use the “For” loop for this task.

# Load project data
df_project = pd.read_csv("project_data.csv")

# Calculate running average and fill missing values
total_sum = 0
count = 0

for i in range(len(df_project)):
    if pd.isna(df_project.loc[i, "Cost"]):  #Check if cost is missing
        if count > 0:
            df_project.loc[i, "Cost"] = total_sum / count  # Replace with running average
    else:
        total_sum += df_project.loc[i, "Cost"]
        count += 1

# Save the updated DataFrame
df_project.to_csv("project_data_cleaned.csv", index=False)

print("Task 2 completed!\nNow Project table is:")
df_project
```

```python
#Task 3: Split the name column in the Employee dataframe into two new columns “First Name”, and “Last Name” and remove the older “name” column.

# Load employee data
df_employee = pd.read_csv("employee_data.csv")

# Split the "Name" column into "First Name" and "Last Name"
df_employee[["First Name", "Last Name"]] = df_employee["Name"].str.split(" ", n=1, expand=True)

# Drop the old "Name" column
df_employee.drop(columns=["Name"], inplace=True)

# Save the updated DataFrame
df_employee.to_csv("employee_data_updated.csv", index=False)

print("Task 3 completed!\nNow Employee is:")
df_employee
```

```python
#Task 4: Join all three dataframes in one single dataframe. Name it “Final”

# Load all three dataframes
df_employee = pd.read_csv("employee_data_updated.csv")
df_seniority = pd.read_csv("seniority_data.csv")
df_project = pd.read_csv("project_data_cleaned.csv")

# Merge Employee and Seniority DataFrames on "ID"
df_merged = pd.merge(df_employee, df_seniority, on="ID")

# Merge the result with Project DataFrame on "ID"
df_final = pd.merge(df_merged, df_project, on="ID")

# Save the final merged DataFrame
df_final.to_csv("final_dataframe.csv", index=False)

print("Task 4 completed! Data merged in 'final_dataframe.csv'.\nNow Table is:")
df_final
```

```python
#Task 5: Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project cost only to employees who have finished the projects.

# Load the final merged dataframe
df_final = pd.read_csv("final_dataframe.csv")

# Calculate Bonus (5% of project cost for "Finished" projects)
df_final["Bonus"] = df_final.apply(lambda row: row["Cost"] * 0.05 if row["Status"] == "Finished" else 0, axis=1)

# Save the updated DataFrame
df_final.to_csv("final_dataframe_with_bonus.csv", index=False)

print("Task 5 completed!\nNow Final table is:")
df_final
```

```python
#Task 6: Demote the designation level by 1, whose projects have status “fail”. Delete the employees record whose designation level is above 4.

# Load the final dataframe with bonus
df_final = pd.read_csv("final_dataframe_with_bonus.csv")

# Demote designation level by 1 for employees with failed projects
df_final.loc[df_final["Status"] == "Failed", "Designation Level"] -= 1

# Remove employees whose designation level exceeds 4
df_final = df_final[df_final["Designation Level"] <= 4]

# Save the updated DataFrame
df_final.to_csv("final_dataframe_demoted.csv", index=False)

print("Task 6 completed!\nNow table is:")
df_final
```

```python
#Task 7: Add “Mr.” and “Mrs.” to the first name column and drop the gender column.

# Load the final dataframe
df_final = pd.read_csv("final_dataframe_demoted.csv")

# Add "Mr." for males and "Mrs." for females
df_final["First Name"] = df_final.apply(lambda row: "Mr. " + row["First Name"] if row["Gender"] == "M" else "Mrs. " + row["First Name"], axis=1)

# Drop the "Gender" column
df_final.drop(columns=["Gender"], inplace=True)

# Save the updated DataFrame
df_final.to_csv("final_dataframe_titles.csv", index=False)

print("Task 7 completed!\nNow table is:")
df_final
```

```python
#Task 8: Promote designation level by 1 for the employees whose age is more than 29 years using IF condition.

# Load the final dataframe
df_final = pd.read_csv("final_dataframe_titles.csv")

# Promote employees (increase designation level) if age > 29
df_final.loc[df_final["Age"] > 29, "Designation Level"] += 1

# Save the updated DataFrame
df_final.to_csv("final_dataframe_promoted.csv", index=False)

print("Task 8 completed!")
df_final
```

```python
#Task 9: Add the cost of all projects for each Employee and save it in new dataframe “TotalProject Cost” with three columns ID, First Name, and Total cost

# Load the final dataframe
df_final = pd.read_csv("final_dataframe_promoted.csv")

# Calculate total project cost per employee
df_total_cost = df_final.groupby(["ID", "First Name"])["Cost"].sum().reset_index()

# Rename the "Cost" column to "Total Cost"
df_total_cost.rename(columns={"Cost": "Total Cost"}, inplace=True)

# Save the new dataframe
df_total_cost.to_csv("total_project_cost.csv", index=False)

print("Task 9 completed! Total project cost per employee calculated.")
df_total_cost
```
